```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("fx_smile_calculus.csv")
for col in data.columns:
    data[col] = data[col].apply(lambda x: str(x))
    data[col] = data[col].apply(lambda x: x.replace(',', '.'))
data
```

```python
#IV in function of log(F/K)
plt.plot(pd.to_numeric(data["log(F/K)"]), pd.to_numeric(data["IV"]))
```

```python
#delta in function of log(F/K)
plt.plot(pd.to_numeric(data["log(F/K)"]), pd.to_numeric(data["delta"]))
```

```python
#IV in function of delta
plt.plot(pd.to_numeric(data["delta"]), pd.to_numeric(data["IV"]))
```
